```python
import pandas as pd
import requests
import time

df = pd.read_csv("/content/drive/MyDrive/olx.csv")

google_api_key = ""

def obter_dados_geocode(endereco, api_key):
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    params = {
        "address": endereco,
        "key": api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        return response.json()
    else:
        return None

def extrair_dados_geocode(json_data):
    if json_data and json_data.get("status") == "OK":
        resultado = json_data["results"][0]
        endereco_formatado = resultado.get("address_components", [])

        cep, bairro, rua, numero = None, None, None, None

        for componente in endereco_formatado:
            if "postal_code" in componente["types"]:
                cep = componente["long_name"]
            elif "sublocality" in componente["types"] or "neighborhood" in componente["types"]:
                bairro = componente["long_name"]
            elif "route" in componente["types"]:
                rua = componente["long_name"]
            elif "street_number" in componente["types"]:
                numero = componente["long_name"]

        latitude = resultado["geometry"]["location"]["lat"]
        longitude = resultado["geometry"]["location"]["lng"]

        return cep, bairro, rua, numero, latitude, longitude
    else:
        return None, None, None, None, None, None

df["cep"] = None
df["bairro"] = None
df["rua"] = None
df["numero"] = None
df["latitude"] = None
df["longitude"] = None

for index, row in df.iterrows():
    endereco = row["descricao_localizacao"]
    print(f"Pesquisando endereço: {endereco}")

    dados_geocode = obter_dados_geocode(endereco, google_api_key)

    cep, bairro, rua, numero, latitude, longitude = extrair_dados_geocode(dados_geocode)

    df.at[index, "cep"] = cep
    df.at[index, "bairro"] = bairro
    df.at[index, "rua"] = rua
    df.at[index, "numero"] = numero
    df.at[index, "latitude"] = latitude
    df.at[index, "longitude"] = longitude

    time.sleep(1)

df.to_csv("/content/drive/MyDrive/olx.csv", index=False)
print("Processo concluído. Dados salvos em '/content/drive/MyDrive/olx.csv'.")
```

```python
import pandas as pd

caminho_df1 = "/content/drive/MyDrive/olx.csv"
caminho_df2 = "/content/drive/MyDrive/Banco_junto_v1.xlsx"

df1 = pd.read_csv(caminho_df1)
df2 = pd.read_excel(caminho_df2)

if 'vendedor' in df1.columns:
    df1 = df1.drop('vendedor', axis=1)

if 'vendedor' in df2.columns:
    df2 = df2.drop('vendedor', axis=1)

df1 = df1.dropna(how='all')
df2 = df2.dropna(how='all')


def compare_rows(row1, row2):
    """Compara linhas com base nas colunas especificadas."""
    cols_to_compare = ['descricao_localizacao', 'longitude', 'latitude', 'valor_area', 'preco']

    for col in cols_to_compare:
         val1 = str(row1[col]).strip().lower() if pd.notna(row1[col]) else ''
         val2 = str(row2[col]).strip().lower() if pd.notna(row2[col]) else ''
         if val1 != val2:
            return False
    return True

merged_rows = []

for _, row1 in df1.iterrows():
    merged_rows.append(row1)

for _, row2 in df2.iterrows():

    add_row=True

    for row1 in merged_rows:
        if compare_rows(row1, row2):
            add_row=False
            break

    if add_row:
        merged_rows.append(row2)

df_merged = pd.DataFrame(merged_rows)

df_merged.to_excel("/content/drive/MyDrive/banco_teste_v1.xlsx", index=False)

print("DataFrames unificados, coluna 'vendedor' excluída, e arquivo 'banco_v1.xlsx' salvo no Google Drive.")
```

```python
import folium
from folium import plugins

caminho_df = "/content/drive/MyDrive/banco_v1.xlsx"

if 'csv' in caminho_df :
  df = pd.read_csv(caminho_df)
elif 'xlsx' in caminho_df:
  df = pd.read_excel(caminho_df)
else:
    print('Nenhum arquivo xlsx ou csv encontrado com o nome')

df=df.dropna(subset=['latitude','longitude'])


latitude_media = df['latitude'].mean()
longitude_media = df['longitude'].mean()

mapa = folium.Map(location=[latitude_media, longitude_media], zoom_start=12)

for index, row in df.iterrows():
    try:
        latitude = float(row['latitude'])
        longitude = float(row['longitude'])
        if pd.notna(latitude) and pd.notna(longitude):
            folium.Marker(
              location=[latitude, longitude],
              popup=f"Localização: {row['descricao_localizacao'] if 'descricao_localizacao' in row else 'Sem Descrição'}<br>Preço: {row['preco'] if 'preco' in row else 'Não disponível'}<br>Tipo: {row['tipo_de_imovel'] if 'tipo_de_imovel' in row else 'Não Definido'}",
              tooltip="Clique para ver detalhes"
             ).add_to(mapa)
    except ValueError:
        print(f"Erro na conversão de latitude/longitude da linha {index}. Verifique seus dados")
    except Exception as e:
        print(f"Erro ao adicionar marcador na linha {index}: {e}")

mapa.save("/content/drive/MyDrive/mapa_caruaru.html")

print("Mapa criado com sucesso e salvo como 'mapa_caruaru.html' no Google Drive.")
```

```python
import pandas as pd
import folium
from folium import plugins

caminho_df = "/content/drive/MyDrive/banco_v1.xlsx"

if 'csv' in caminho_df :
  df = pd.read_csv(caminho_df)
elif 'xlsx' in caminho_df:
  df = pd.read_excel(caminho_df)
else:
    print('Nenhum arquivo xlsx ou csv encontrado com o nome')

df=df.dropna(subset=['latitude','longitude'])

heat_data = [[row['latitude'], row['longitude']] for index, row in df.iterrows() if pd.notna(row['latitude']) and pd.notna(row['longitude'])]

latitude_media = df['latitude'].mean()
longitude_media = df['longitude'].mean()

mapa = folium.Map(location=[latitude_media, longitude_media], zoom_start=12)

plugins.HeatMap(heat_data).add_to(mapa)

mapa.save("/content/drive/MyDrive/mapa_calor_caruaru.html")

print("Mapa de calor criado com sucesso e salvo como 'mapa_calor_caruaru.html' no Google Drive.")
```
